```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from IPython.display import display, clear_output
from collections import Counter
```

# Data

```python
df = pd.read_csv('vehicles.csv', low_memory=False)
df.head(10)
```

# Simple Regression

```python
car_model = []
statistics = [[]]
for i in range(len(df)):
    new = 1
    # search models
    for j in range(len(car_model)):
        if df['make'][i] == car_model[j]:
            statistics[j].append((df['highway08'][i],df['co2TailpipeGpm'][i]))
            new = 0
    # new model
    if new == 1:
        car_model.append(df['make'][i])
        statistics.append([(df['highway08'][i],df['co2TailpipeGpm'][i])])
```

```python
print_model = ["Ferrari", "Dodge", "Toyota"]
print_statistics = [[]]

for i in range(len(car_model)):
    if car_model[i] == "Ferrari":
        print_statistics[0] = statistics[i] 
    elif car_model[i] == "Dodge":
        print_statistics.append(statistics[i])     
    elif car_model[i] == "Toyota":
        print_statistics.append(statistics[i]) 
        
for i in range(len(print_model)):
    x,y = zip(*print_statistics[i])
    plt.scatter(x, y)
    
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.plot(x,p(x))
    
plt.xlabel("Highway MPG")
plt.ylabel("Tailpipe CO2 Emission (GPM)")
plt.legend(print_model)
```
